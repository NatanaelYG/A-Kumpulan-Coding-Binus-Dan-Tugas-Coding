```python
import numpy as np

#ruas kiri
xSoal=[
    [
      [4, 2, -1],
      [1, -5, 2],
      [2, -1, -4]
    ],
    [
      [3, 4, 5],
      [-3, 7, -4],
      [1, -4, -2]
    ],
    [
      [9, -2, 3, 2],
      [2, 8, -2, 3],
      [-3, 2, 11, -4],
      [-2, 3, 2, 10]
    ]
]

#ruas kanan
ySoal=[
    [41, -10, 1],
    [34, -32, 62],
    [55, -14, 12, -21]
]


def checkDiag(x,y):
    xArr = np.array(x)

    diag = np.diag(np.abs(xArr))
    sumNoDiag = np.sum(np.abs(xArr), axis = 1) - diag

    if np.all(diag > sumNoDiag) :
        print('diagonally dominant\n')
        gaussSeidel(x,y)
    else :
        print('not diagonally dominant\n')

def gaussSeidel(x, y, eps = 0.022):
    arrKiri = np.array(x)
    arrKanan = np.array(y)
    # converged = False

    x_old = np.zeros(arrKiri.shape[0])
    diagonal = np.diag(arrKiri)
    arrKiri = -arrKiri
    np.fill_diagonal(arrKiri, 0)

    for i in range(15):
        x_new = np.array(x_old)

        for j, row in enumerate(arrKiri):
            x_new[j] = (arrKanan[j] + np.dot(row, x_new)) / diagonal[j]

        print(f'iterasi ke - {i+1} x = {x_new}')

        distance = np.sqrt(np.dot((x_new - x_old), (x_new - x_old)))

        if distance < eps:
            print('converged\n')
            # converged = True
            break
        
        x_old = x_new

        # if converged == False:
        #     print('not converged')
```

```python
for i, (ruasKiri, ruasKanan) in enumerate (zip(xSoal, ySoal)):
    print(ruasKiri)
    print(ruasKanan)
    checkDiag(ruasKiri, ruasKanan)
```
