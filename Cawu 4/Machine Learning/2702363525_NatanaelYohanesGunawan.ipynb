```python
import pandas as pd
ds = pd.read_csv('titanic.csv')
ds.head()
```

```python
ds[['Survived','Pclass','Sex']].groupby('Survived').value_counts().unstack().plot.bar()
```

Pada Visualisasi di atas Dapat dilihat
1. Class yang paling banyak bertahan hidup adalah kelas 3 dengan jumlah lebih banyak dari kelas yang lainnya.
2. Dan orang yang paling banyak bertahan hidup berjenis kelamin pria dibanding dengan wanita yang jauh lebih rendah pada kelas 3. 

```python
ds[['Age','Sex']].plot.hist(by='Sex')
```

3. Rata-rata umur yang memiliki populasi paling banyak adalah di antara umur 20 sampai 30 untuk pria dan 15 sampai 32 untuk wanita 
4. Dari data di atas bahwa populasi Pria lebih banyak daripada wanita dimana wanita memiliki frequency paling banyak adalah di bawah 50 dan untuk Pria jauh lebih banyak dimana di atas seratus yang dimana populasi pria lebih banyak 2 kali lipat dibanding dengan wanita

```python
ds['color'] = ds.Sex.map({'male':'blue','female':'green'})

ds[['Age', 'Fare','color']].plot.scatter(x='Age',y='Fare',c='color')
```

5. Pada umur antara 30 sampai 40 memiliki beberapa titik yang membayar dengan  ongkos yang mahal
6. Bisa dilihat pada data bahwa rata-rata pelanggan paling banyak membayar ongkos di bawah 100
