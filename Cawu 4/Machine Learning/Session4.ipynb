```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('titanic.csv')

df.tail()
```

```python
df.info()
```

```python
# Drop fising value
#df.dropna(inplace=True)
#df.info()
```

```python
# Remove pising value
df = df.drop(columns=["Age", "Embarked", "Cabin"])
df.info()
```

```python
# Remove non numeric data
# Remove incomplete data
df = df.drop(columns=["Name","Sex","Ticket"])
df.info()
```

```python
# Remove irrelevant data
df = df.drop(columns=["PassengerId"])
df.info()
```

```python
# Make classifier using logic regression

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

x = df.drop(columns=["Survived"])
y = df["Survived"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=26)

model = LogisticRegression(solver="sag")
model.fit(x_train, y_train)
```

```python
#meansure model performence using accuracy, preccision, recall, f1_score, and auc-roc

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

y_pred = model.predict(x_test)

print("Accuracy: ", accuracy_score(y_test, y_pred))
print("Precision: ", precision_score(y_test, y_pred))
print("Recall: ", recall_score(y_test, y_pred))
print("F1: ", f1_score(y_test, y_pred))
print("ROC: ", roc_auc_score(y_test, y_pred))
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
df = pd.read_csv('titanic.csv')

df.drop(columns=["Age", "Embarked", "Cabin"], inplace=True)
df.drop(columns=["Name"], inplace=True)
df.drop(columns=["PassengerId"], inplace=True)

df.info()
```

```python
# Label Encoding
# Encode ticket using label encoder
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df["Ticket"] = le.fit_transform(df["Ticket"])
df.info()
```

```python
# One hot encoding for sex using pandas get dummies

df = pd.get_dummies(df, columns=["Sex"])
df.head()
```

```python
# Modeling
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

x = df.drop(columns=["Survived"])
y = df["Survived"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=48)

model = LogisticRegression(solver="lbfgs")
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)

print(classification_report(y_test, y_pred))
```

```python
def Pclass(class_num):
    if class_num == 1:
        return 100
    if class_num == 2:
        return 20
    elif class_num == 3:
        return 10
    
df['Pclass'] = df['Pclass'].apply(Pclass)
df.head()
```

```python
random_state_values = range(1, 100) 
best_f1_score = 0 
best_random_state = None 

for random_state in random_state_values: 
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=random_state) 
    model = LogisticRegression(solver="lbfgs", max_iter=100) 
    model.fit(x_train, y_train) 
    y_pred = model.predict(x_test) 
    f1 = f1_score(y_test, y_pred, average='macro') 
    if f1 > best_f1_score: 
        best_f1_score = f1 
        best_random_state = random_state 

# Menggunakan random_state terbaik 

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=best_random_state) 
model = LogisticRegression(solver="lbfgs", max_iter=100) 
model.fit(x_train, y_train) 
y_pred = model.predict(x_test) 
print(f"Best random_state: {best_random_state}") 
print(classification_report(y_test, y_pred))
```
