```python
#split data
from sklearn.model_selection import train_test_split

x = df.drop('Class', axis = 1)
y = df['Class']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.5, random_state = 26)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
y_test.value_counts()
```

```python
# gunakan pipe line dimana preproccsing standard scalle lalu modeling

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import RobustScaler


pipe = Pipeline([
    ('scaler', RobustScaler()),
    ('logreg', LogisticRegression(solver='liblinear'))
])


pipe.fit(x_train, y_train)


# k fold cross
scores = cross_val_score(pipe, x, y, cv=5)
print(f'Mean Score : {scores.mean()}')
print(f'Std Score : {scores.std()}')
```

```python
# cel classigication report dengan data test

from sklearn.metrics import classification_report

y_pred = pipe.predict(x_test)
print(classification_report(y_test, y_pred))
```
