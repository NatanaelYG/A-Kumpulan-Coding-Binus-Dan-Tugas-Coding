```python
from Bio.Seq import Seq

SeqA = Seq('AGCTTGCAGCGTCCGTTAGCTCGAGTCCAGGACGTTAGTCCTGCAGTC')
SeqB = Seq('CAGTAAGTTGCCGTTAGCGCGTAGTGCCAGTAAGCGGCTCGTTAGTGG')
```

```python
from Bio.SeqUtils import molecular_weight

print(molecular_weight(SeqA))
print(molecular_weight(SeqB))
```

```python
def wallance(SeqA):
    a_count = SeqA.count('A')
    g_count = SeqA.count('G')
    c_count = SeqA.count('C')
    t_count = SeqA.count('T')

    if len(SeqA) < 14:
        return (a_count + t_count) * 2 + (g_count + c_count) * 4
    else:
        return round(64.9 + 41 * (g_count + c_count - 16.4)/ len(SeqA))
print(wallance(SeqA))
```

```python
def wallance(SeqB):
    a_count = SeqB.count('A')
    g_count = SeqB.count('G')
    c_count = SeqB.count('C')
    t_count = SeqB.count('T')

    if len(SeqB) < 14:
        return (a_count + t_count) * 2 + (g_count + c_count) * 4
    else:
        return round(64.9 + 41 * (g_count + c_count - 16.4)/ len(SeqB))
print(wallance(SeqB))
```

```python
import matplotlib.pyplot as plt

plt.bar(['Seq A', 'Seq B'], [wallance(SeqA), wallance(SeqB)])
plt.figure(figsize=(5,3))
plt.show()
```

```python
from Bio.SeqUtils import gc_fraction

print(f'GC: {gc_fraction(SeqA)}')

print(f'AT: {1 - gc_fraction(SeqA)}')
```

```python
print(f'GC: {gc_fraction(SeqB)}')

print(f'AT: {1 - gc_fraction(SeqB)}')
```

```python
def gc_content(SeqA):
    return (SeqA.count('G') + SeqA.count('C')) / len(SeqA) * 100

def at_content(SeqA):
    return (SeqA.count('A') + SeqA.count('T')) / len(SeqA) * 100

print(f'GC: {gc_content(SeqA)}')
print(f'AT: {at_content(SeqA)}')
```

```python
def gc_content(SeqB):
    return (SeqB.count('G') + SeqB.count('C')) / len(SeqB) * 100

def at_content(SeqB):
    return (SeqB.count('A') + SeqB.count('T')) / len(SeqB) * 100

print(f'GC: {gc_content(SeqB)}')
print(f'AT: {at_content(SeqB)}')
```

```python
gc_A = gc_content(SeqA)
gc_B = gc_content(SeqB)

print('GC seq A', gc_A)
print('GC seq B', gc_B)
```

```python
at_A = at_content(SeqA)
at_B = at_content(SeqB)

print('AT seq A', at_A)
print('AT seq B', at_B)
```

```python
plt.pie([gc_A, at_A], autopct='%.2f%%')
plt.legend(['GC', 'AT'])
plt.title('Sequence A')
plt.figure(figsize=(5,2))
plt.show()

plt.pie([gc_B, at_B], autopct='%.2f%%')
plt.legend(['GC', 'AT'])
plt.title('Sequence B')
plt.show()
```
